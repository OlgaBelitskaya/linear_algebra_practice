```python
# sage
i,j,k = vector(SR,[1,0,0]), vector(SR,[0,1,0]), vector(SR,[0,0,1])
V = vector(SR,[2,3,2])
2 * i + 3 * j + 2 * k == V
```

## 4. Пример типового расчета.

```python
%%latex
$\overrightarrow{C} (3,4,1); \ \overrightarrow{D} (1,2,1) \\
\overrightarrow{A} + \overrightarrow{B} = \overrightarrow{C}; \ 
\overrightarrow{A} \perp \overrightarrow{D}; \ 
\overrightarrow{B} \parallel \overrightarrow{D} \\
\overrightarrow{A}, \ \overrightarrow{B} = \ ?$
```

```python
# sage
a1,a2,a3,b1,b2,b3,k = var('a1 a2 a3 b1 b2 b3 k')
A, B = vector(SR,[a1,a2,a3]), vector(SR,[b1,b2,b3])
C, D = vector(SR,[3,4,1]), vector(SR,[1,2,1])
A, B, C, D
```

```python
solve([(A+B-C)[0]==0,(A+B-C)[1]==0,(A+B-C)[2]==0, # A + B = C
       A.dot_product(D)==0, # A ⊥ C
       (k*D-B)[0]==0,(k*D-B)[1]==0,(k*D-B)[2]==0], # B || D
      a1,a2,a3,b1,b2,b3,k)
```

```python
a=arrow3d((0,0,0),(1,0,-1),color='#3636ff',width=2)+text3d('A',(1+r,0,-1))
b=arrow3d((0,0,0),(2,4,2),color='#ff3636',width=2)+text3d('B',(2+r,4,2))
c=arrow3d((0,0,0),(3,4,1),color='#ff36ff',width=2)+text3d('C',(3+r,4,1))
d=arrow3d((0,0,0),(1,2,1),color='#36ff56',width=2)+text3d('D',(1+r,2,1))
e=line([(1,0,-1),(3,4,1),(2,4,2)])
(a+b+c+d+e).show(aspect_ratio=(3,2,3))
```
