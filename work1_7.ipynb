## 4. –ü—Ä–∏–º–µ—Ä —Ç–∏–ø–æ–≤–æ–≥–æ —Ä–∞—Å—á–µ—Ç–∞.

```python
%%html
<p style="color:slategray;">
$\mathscr{\overrightarrow{C} (3,4,1); \ \overrightarrow{D} (1,2,1) \\
\overrightarrow{A} + \overrightarrow{B} = \overrightarrow{C}; \\ 
\overrightarrow{A} \perp \overrightarrow{D} \ 
(\overrightarrow{A} \cdot \overrightarrow{D} = 0) ;\\ 
\overrightarrow{B} \parallel \overrightarrow{D} \ 
(\overrightarrow{D} * k = \overrightarrow{B});\\
\overrightarrow{A}(a_1, a_2, a_3), \ 
\overrightarrow{B} (b_1, b_2, b_3), \ k = \ ?}$</p>
```

```python
var('k')
A,B=vector(SR,[a1,a2,a3]),vector(SR,[b1,b2,b3])
C,D=vector(SR,[3,4,1]),vector(SR,[1,2,1])
solution=\
solve([(A+B-C)[0]==0,(A+B-C)[1]==0,(A+B-C)[2]==0,#A+B=C
       A.dot_product(D)==0,#A‚ä•D
       (k*D-B)[0]==0,(k*D-B)[1]==0,(k*D-B)[2]==0],#B||D
      a1,a2,a3,b1,b2,b3,k)
A=vector(SR,[solution[0][0].rhs(),
             solution[0][1].rhs(),
             solution[0][2].rhs()])
B=vector(SR,[solution[0][3].rhs(),
             solution[0][4].rhs(),
             solution[0][5].rhs()])
solution
```

```python
vs=[A,B,C,D]; vl=['A','B','C','D']
colors=['#3636ff','#36ff36','#ff3636','#ff36ff']
sum([arrow3d([0,0,0],vs[n],color=colors[n],width=3)+\
text3d(vl[n],vs[n]*1.1) for n in [0..3]]).show()
```

```python
def equations(v):
    a1,a2,a3,b1,b2,b3,k=v
    f1=a1+b1-3; f2=a2+b2-4; f3=a3+b3-1 #A+B=C
    f4=a1*1+a2*2+a3*1 #A‚ä•D
    f5=k*1-b1; f6=k*2-b2; f7=k*1-b3 #B||D 
    return (f1,f2,f3,f4,f5,f6,f7)
a1,a2,a3,b1,b2,b3,k=\
round(fsolve(equations,(0,0,0,0,0,0,0)))
[(a1,a2,a3),(b1,b2,b3),k]
```

```python
a1,a2,a3,b1,b2,b3,k=sympy.symbols('a1 a2 a3 b1 b2 b3 k')
A,B=sympy.Matrix([a1,a2,a3]),sympy.Matrix([b1,b2,b3])
C,D=sympy.Matrix([3,4,1]),sympy.Matrix([1,2,1])
a1,a2,a3,b1,b2,b3,k=\
sympy.nsolve(((A+B-C)[0],(A+B-C)[1],(A+B-C)[2],#A+B=C
              A.dot(D),#A‚ä•D
              (k*D-B)[0],(k*D-B)[1],(k*D-B)[2]),#B||D
              (a1,a2,a3,b1,b2,b3,k),(0,0,0,0,0,0,0))
[(a1,a2,a3),(b1,b2,b3),k]
```

```python
%%r
check_decision<-function(x){
    d<-c(x[1]+x[4]-3==0,#A+B=C
         x[2]+x[5]-4==0, 
         x[3]+x[6]-1==0, 
         x[1]*1+x[2]*2+x[3]*1==0,#A‚ä•D
         x[7]*1-x[4]==0,#B||D 
         x[7]*2-x[5]==0, 
         x[7]*1-x[6]==0)
    sum(d)==length(d)}
check_decision(c(1,0,-1,2,4,2,2))
```

```python
%%r
library(pracma)
problem_decision<-function(x){
    c(x[1]+x[4]-3, # A + B = C
      x[2]+x[5]-4, 
      x[3]+x[6]-1, 
      x[1]*1+x[2]*2+x[3]*1, # A ‚ä• D
      x[7]*1-x[4], # B || D 
      x[7]*2-x[5], 
      x[7]*1-x[6])}
print('a1 a2 a3 b1 b2 b2 k')
round(fsolve(problem_decision,c(0,0,0,0,0,0,0))$x)
```
