```
k = 0.1
a = arrow3d((0,0,0),(1,0,-1),color='#3636ff',width=3)+text3d('A',(1+k,0,-1))
b = arrow3d((0,0,0),(2,4,2),color='#ff3636',width=3)+text3d('B',(2+k,4,2))
c = arrow3d((0,0,0),(3,4,1),color='#ff36ff',width=3)+text3d('C',(3+k,4,1))
d = arrow3d((0,0,0),(1,2,1),color='#36ff56',width=3)+text3d('D',(1+k,2,1))
(a+b+c+d).show(aspect_ratio=(4,2,3))
```
