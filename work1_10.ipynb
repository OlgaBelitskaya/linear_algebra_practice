## 1. Эллипс

Окружность

С центром в точке $O (x_0;y_0;z_0)$ и радиусом $r$

$(x - x_0)^2 + (y - y_0)^2 = r^2$

Параметрические

$x = r * \cos t + x_0, y = r * \sin t + y_0, 0 \leq t < 2\pi$

Эллипс

С центром в точке $O (x_0;y_0;z_0)$ и полуосями $a,b$

$\frac{(x - x_0)^2}{a^2} + \frac{(y - y_0)^2}{b^2} = 1$

```python
# sage
x,y,x0,y0,a,b,r,t = var('x,y,x0,y0,a,b,r,t')

eq1 = (x-x0)**2+(y-y0)**2-r**2==0
eq2 = x==r*cos(t)+x0
eq3 = y==r*sin(t)+y0
eq4 = (x-x0)**2/a**2+(y-y0)**2/b**2-1==0
```

```python
solve([eq1, x0==1, y0==-1, r==2], x0,y0,r,x,y)
```

```python
solve([eq2, eq3, x0==-1, y0==1, r==3], x0,y0,r,t,x,y)
```

```python
solve([eq4, x0==1, y0==-1, a==1, b==3], x0,y0,a,b,x,y)
```

```python
g = Graphics()
p1 = implicit_plot(lambda x,y: (x-1)**2+(y+1)**2-2**2, (-2,4), (-4,2), color='#3636ff')
t1 = text('(x-1)^2+(y+1)^2=2^2', (1,1.5), color='#3636ff')
p2 = parametric_plot((3*cos(t)-1, 3*sin(t)+1), (t,0,2*pi), color='#ff3636', gridlines=True)
t2 = text('x=3cos(t)-1, y=3sin(t)+1', (-1,4.5), color='#ff3636')
p3 = implicit_plot(lambda x,y: (x-1)**2/1**2+(y+1)**2/3**2-1, (-1,3), (-5,3), color='#36ff36')
t3 = text('(x-1)^2/1^2+(y+1)^2/3^2=1', (1,-4.5), color='#36ff36')
g = g+p1+t1+p2+t2+p3+t3
g.show()
```

```python
import sympy
sympy.init_printing(use_unicode=True)
x,y,x0,y0,a,b,r,t = sympy.symbols('x,y,x0,y0,a,b,r,t')

eq1 = (x-x0)**2+(y-y0)**2-r**2
eq2 = x-r*cos(t)+x0
eq3 = y-r*sin(t)+y0
eq4 = (x-x0)**2/a**2+(y-y0)**2/b**2-1
```

```python
sympy.solve([eq1, x0-1, y0+1, r-2], x0,y0,r,y)
```

```python
sympy.solve([eq2, eq3, x0+1, y0-1, r-3], x0,y0,r,x,y)
```

```python
sympy.solve([eq4, x0-1, y0+1, a-1, b-3], x0,y0,a,b,y)
```
