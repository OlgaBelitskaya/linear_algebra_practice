## 1. –≠–ª–ª–∏–ø—Å –∏ –æ–∫—Ä—É–∂–Ω–æ—Å—Ç—å

#### –û–∫—Ä—É–∂–Ω–æ—Å—Ç—å

–° —Ü–µ–Ω—Ç—Ä–æ–º –≤ —Ç–æ—á–∫–µ $A_0 (x_0;y_0)$ –∏ —Ä–∞–¥–∏—É—Å–æ–º $r$

$(x - x_0)^2 + (y - y_0)^2 = r^2$

–ü–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–µ

$x = r * \cos t + x_0, y = r * \sin t + y_0, 0 \leq t < 2\pi$

–û–±—â–µ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ

$a * x^2 + a * y^2 + d * x + e * y + f = 0$

—Å —Ü–µ–Ω—Ç—Ä–æ–º –≤ —Ç–æ—á–∫–µ $\bigg(x_0 = -\frac{d}{2*a}; y_0 = -\frac{e}{2*a}\bigg)$ –∏ —Ä–∞–¥–∏—É—Å–æ–º $r = \frac{\sqrt{d^2 + e^2 - 4 * a * f}}{2 * |a|}$

–ß–µ—Ä–µ–∑ —Ç—Ä–∏ —Ç–æ—á–∫–∏ $A_1(x_1; y_1), \ A_2(x_2; y_2), \ A_3(x_3; y_3)$

$\begin{vmatrix}
x^2+y^2 & x & y & 1 \\
x_1^2+y_1^2 & x_1 & y_1 & 1 \\
x_2^2+y_2^2 & x_2 & y_2 & 1 \\
x_3^2+y_3^2 & x_3 & y_3 & 1 
\end{vmatrix} = 0$

–î–ª–∏–Ω–∞ –ø–µ—Ä–∏–º–µ—Ç—Ä–∞

$L = 2 * \pi * r$

–ü–ª–æ—â–∞–¥—å 

$S =\pi * r^2$

#### –≠–ª–ª–∏–ø—Å

–° —Ü–µ–Ω—Ç—Ä–æ–º –≤ —Ç–æ—á–∫–µ $A_0 (x_0;y_0)$ –∏ –ø–æ–ª—É–æ—Å—è–º–∏ $p,q \ (p > q)$

–ö–∞–Ω–æ–Ω–∏—á–µ—Å–∫–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ

$\frac{(x - x_0)^2}{p^2} + \frac{(y - y_0)^2}{q^2} = 1$

–≠–∫—Å—Ü–µ–Ω—Ç—Ä–∏—Å–∏—Ç–µ—Ç

$\epsilon = \frac{\sqrt{p^2 - q^2}}{p}$

–§–æ–∫—É—Å—ã $F_1(x_0-\sqrt{p^2 - q^2};y_0), \ F_2(x_0+\sqrt{p^2 - q^2};y_0)$

$F_1M + F_2M = 2*p, \ \forall M \in ellipse$

–ü–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–µ

$x = p * \cos t + x_0, y = q * \sin t + y_0, 0 \leq t < 2\pi$

–û–±—â–µ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ

$a * x^2 + b * x * y + c * y^2 + d * x + e * y + f = 0$

$b^2 - 4*a*c < 0$

–° –ø–æ–ª—É–æ—Å—è–º–∏, –ø–∞—Ä–∞–ª–ª–µ–ª—å–Ω—ã–º–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω—ã–º –æ—Å—è–º

$a * x^2 + —Å * y^2 + d * x + e * y + f = 0$

$a*c > 0$

–î–ª–∏–Ω–∞ –ø–µ—Ä–∏–º–µ—Ç—Ä–∞

$L \approx \pi * \sqrt{2 * (p^2 + q^2)}$

–ü–ª–æ—â–∞–¥—å 

$S =\pi * p * q$

–ö–∞—Å–∞—Ç–µ–ª—å–Ω–∞—è –≤ —Ç–æ—á–∫–µ $M (x_M; y_M)$

$\frac{(x - x_0)*(x_M - x_0)}{p^2} + \frac{(y - y_0)*(y_M - y_0)}{q^2} = 1$

```python
var('x,y,x0,y0,a,b,c,d,e,f,p,q,r,t')
eq1=[(x-x0)**2+(y-y0)**2-r**2==0,
     x==r*cos(t)+x0,y==r*sin(t)+y0,
     (x-x0)**2/p**2+(y-y0)**2/q**2-1==0,
     a*x**2+b*x*y+c*y**2+d*x+e*y+f==0]
```

```python
solutions1=[solve([eq1[0],x0==1,y0==-1,r==2],x0,y0,r,y),
            solve([eq1[1],eq1[2],
                   x0==-1,y0==1,r==3],x0,y0,r,x,y),
            solve([eq1[3],x0==1,y0==-1,
                   p==5,q==4],x0,y0,p,q,y),
            solve([eq1[4],a==1,b==1,c==1,
                   d==2,e==6,f==3],a,b,c,d,e,f,y)]
for s in solutions1: pretty_print(s)
```

## 2. –ì–∏–ø–µ—Ä–±–æ–ª–∞
–° —Ü–µ–Ω—Ç—Ä–æ–º –≤ —Ç–æ—á–∫–µ $A_0 (x_0;y_0)$ –∏ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ–π –∏ –º–Ω–∏–º–æ–π –ø–æ–ª—É–æ—Å—è–º–∏ $p,q$

–ö–∞–Ω–æ–Ω–∏—á–µ—Å–∫–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ

$\frac{(x - x_0)^2}{p^2} - \frac{(y - y_0)^2}{q^2} = 1$

–≠–∫—Å—Ü–µ–Ω—Ç—Ä–∏—Å–∏—Ç–µ—Ç

$\epsilon = \frac{\sqrt{p^2 + q^2}}{p}$

–£—Ä–∞–≤–Ω–µ–Ω–∏—è –∞—Å–∏–º–ø—Ç–æ—Ç –≥–∏–ø–µ—Ä–±–æ–ª—ã  

$y = \pm \frac{q}{p} * (x - x_0) + y_0$

–§–æ–∫—É—Å—ã $F_1(x_0-\sqrt{p^2 + q^2}; y_0), \ F_2(x_0+\sqrt{p^2 + q^2}; y_0)$

$|F_1M - F_2M| = 2*p, \ \forall M \in hyperbole$

–ü–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–µ (–ø—Ä–∞–≤–∞—è –≤–µ—Ç–≤—å)

$x = p * \cosh t + x_0, y = q * \sinh t + y_0, 0 \leq t < 2\pi$

–û–±—â–µ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ

$a * x^2 + b * x * y + —Å * y^2 + d * x + e * y + f = 0$

$b^2 - 4*a*c > 0$

–° –ø–æ–ª—É–æ—Å—è–º–∏, –ø–∞—Ä–∞–ª–ª–µ–ª—å–Ω—ã–º–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω—ã–º –æ—Å—è–º

$a * x^2 + —Å * y^2 + d * x + e * y + f = 0$

$a*c < 0$

–ö–∞—Å–∞—Ç–µ–ª—å–Ω–∞—è –≤ —Ç–æ—á–∫–µ $M (x_M; y_M)$

$\frac{(x - x_0)*(x_M - x_0)}{p^2} - \frac{(y - y_0)*(y_M - y_0)}{q^2} = 1$
