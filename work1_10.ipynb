## 1. Эллипс и окружность

#### Окружность

С центром в точке $A_0 (x_0;y_0;z_0)$ и радиусом $r$

$(x - x_0)^2 + (y - y_0)^2 = r^2$

Параметрические

$x = r * \cos t + x_0, y = r * \sin t + y_0, 0 \leq t < 2\pi$

Общее уравнение

$a * x^2 + a * y^2 + d * x + e * y + f = 0$

с центром в точке $\bigg(x_0 = -\frac{d}{2*a}; y_0 = -\frac{e}{2*a}\bigg)$ и радиусом $r = \frac{\sqrt{d^2 + e^2 - 4 * a * f}}{2 * |a|}$

Через три точки $A_1(x_1; y_1), \ A_2(x_2; y_2), \ A_3(x_3; y_3)$

$\begin{vmatrix}
x^2+y^2 & x & y & 1 \\
x_1^2+y_1^2 & x_1 & y_1 & 1 \\
x_2^2+y_2^2 & x_2 & y_2 & 1 \\
x_3^2+y_3^2 & x_3 & y_3 & 1 
\end{vmatrix} = 0$

Длина периметра

$L = 2 * \pi * r$

Площадь 

$S =\pi * r^2$

#### Эллипс

С центром в точке $A_0 (x_0;y_0;z_0)$ и полуосями $p,q \ (p > q)$

Каноническое уравнение

$\frac{(x - x_0)^2}{p^2} + \frac{(y - y_0)^2}{q^2} = 1$

Эксцентриситет

$\epsilon = \frac{\sqrt{p^2 - q^2}}{p}$

Фокусы $F_1(x_0-\sqrt{p^2 - q^2};y_0), \ F_2(x_0+\sqrt{p^2 - q^2};y_0)$

$F_1M + F_2M = 2*p, \ \forall M \in ellipse$

Параметрические

$x = p * \cos t + x_0, y = q * \sin t + y_0, 0 \leq t < 2\pi$

Общее уравнение

$a * x^2 + b * x * y + с * y^2 + d * x + e * y + f = 0$

$b^2 - 4*a*c < 0$

С полуосями, параллельными координатным осям

$a * x^2 + с * y^2 + d * x + e * y + f = 0$

$a*c > 0$

Длина периметра

$L \approx \pi * \sqrt{2 * (p^2 + q^2)}$

Площадь 

$S =\pi * p * q$

Касательная к эллипсу в точке $M (x_M; y_M)$

$\frac{(x - x_0)*(x_M - x_0)}{p^2} + \frac{(y - y_0)*(y_M - y_0)}{q^2} = 1$

```python
# sage
x,y,x0,y0,a,b,c,d,e,f,p,q,r,t = var('x,y,x0,y0,a,b,c,d,e,f,p,q,r,t')

eq1 = (x-x0)**2+(y-y0)**2-r**2==0
eq2 = x==r*cos(t)+x0
eq3 = y==r*sin(t)+y0
eq4 = (x-x0)**2/p**2+(y-y0)**2/q**2-1==0
eq5 = a*x**2+b*x*y+c*y**2+d*x+e*y+f==0
```

```python
solve([eq1, x0==1, y0==-1, r==2], x0,y0,r,y)
```

```python
solve([eq2, eq3, x0==-1, y0==1, r==3], x0,y0,r,x,y)
```

```python
solve([eq4, x0==1, y0==-1, p==5, q==4], x0,y0,p,q,y)
```

```python
solve([eq5, a==1, b==1, c==1, d==2, e==6, f==3], a,b,c,d,e,f,y)
```

## 2. Гипербола
С центром в точке $A_0 (x_0;y_0;z_0)$ и действительной и мнимой полуосями $p,q$

Каноническое уравнение

$\frac{(x - x_0)^2}{p^2} - \frac{(y - y_0)^2}{q^2} = 1$

Эксцентриситет

$\epsilon = \frac{\sqrt{p^2 + q^2}}{p}$

Уравнения асимптот гиперболы  

$y = \pm \frac{q}{p} * (x - x_0) + y_0$

Фокусы $F_1(x_0-\sqrt{p^2 + q^2}; y_0), \ F_2(x_0+\sqrt{p^2 + q^2}; y_0)$

$|F_1M - F_2M| = 2*p, \ \forall M \in hyperbole$

Параметрические (правая ветвь)

$x = p * \cosh t + x_0, y = q * \sinh t + y_0, 0 \leq t < 2\pi$

Общее уравнение

$a * x^2 + b * x * y + с * y^2 + d * x + e * y + f = 0$

$b^2 - 4*a*c > 0$

С полуосями, параллельными координатным осям

$a * x^2 + с * y^2 + d * x + e * y + f = 0$

$a*c < 0$

```python
# sage
x,y,x0,y0,a,b,c,d,e,f,p,q,t = var('x,y,x0,y0,a,b,c,d,e,f,p,q,t')

eq1 = (x-x0)**2/p**2-(y-y0)**2/q**2-1==0
eq2 = x==p*cosh(t)+x0
eq3 = y==q*sinh(t)+y0
eq4 = a*x**2+b*x*y+c*y**2+d*x+e*y+f==0
```

```python
solve([eq1, x0==1, y0==-1, p==3, q==4], x0,y0,p,q,y)
```

```python
solve([eq2, eq3, x0==-1, y0==1, p==3, q==4], x0,y0,p,q,x,y)
```

```python
solve([eq4, a==1, b==1, c==-1, d==2, e==6, f==3], a,b,c,d,e,f,y)
```

## 3. Парабола

Каноническое с вершиной $A_0(x_0; y_0)$

$(y - y_0)^2 = 2 * p * (x - x_0)$

Фокус $F(x_0 + \frac{p}{2}; y_0)$

Общее уравнение

$a * x^2 + b * x * y + с * y^2 + d * x + e * y + f = 0$

$b^2 - 4*a*c = 0$

Ось симметрии параллельна оси $Oy$

$a * x^2 + d * x + e * y + f = 0$

Ось симметрии параллельна оси $Ox$

$c * y^2 + d * x + e * y + f = 0$

```python
# sage
x,y,x0,y0,a,b,c,d,e,f,p = var('x,y,x0,y0,a,b,c,d,e,f,p')

eq1 = (y-y0)**2==2*p*(x-x0)
eq2 = a*x**2+b*x*y+c*y**2+d*x+e*y+f==0
eq3 = a*x**2+d*x+e*y+f==0
```

```python
solve([eq1, x0==1, y0==-1, p==3], x0,y0,p,y)
```

```python
solve([eq2, a==1, b==-4, c==4, d==1, e==4, f==3], a,b,c,d,e,f,y)
```

```python
solve([eq3, a==2, d==4, e==1, f==-7], a,d,e,f,y)
```
